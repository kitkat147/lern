```python
import tkinter as tk
import random

root = tk.Tk() #creates a window

root.geometry("600x600") #dimensions of the window created
root.title("can you do the super bag?") #title of the window

#initial coordinates of basket
x_coord_of_basket = 6*50

basket = tk.Label(root, text="[___]", anchor = "center", font=('Arial', 18)) #parent object, what to put into, format
basket.place(x = x_coord_of_basket - 15, y = 20*25)

def go_left():
    global x_coord_of_basket
    if x_coord_of_basket > 3*50:
        x_coord_of_basket = x_coord_of_basket - 50
        basket.place(x = x_coord_of_basket - 15, y = 20*25)
    return

def go_right():
    global x_coord_of_basket
    if x_coord_of_basket < 9*50:
        x_coord_of_basket = x_coord_of_basket + 50
        basket.place(x = x_coord_of_basket - 15, y = 20*25)
    return

def move_basket(keyboardpress):
    #print(x_coord_of_basket)
    #print(keyboardpress) #can see each individual keypresses and the change in state, x y is the position of mouse when keypress happens?
    #print(event.keysym) #show the specific properties
    #print(event.state)
    if keyboardpress.keysym == "a" or keyboardpress.keysym == "Left":
        go_left() #activating/calling go_left
    elif keyboardpress.keysym == "d" or keyboardpress.keysym == "Right":
        go_right() #activating/calling

    elif keyboardpress.keysym == "Escape":
        if running1 == True:
            running_switch_1()
        if running2 == True:
            running_switch_2()


def end_the_game():
    global running1, running2
    def restart_game():
        global y_object, x_object, object_value, answer, question, score, x_coord_of_basket, num_lives, basket_values_ls
        endscreen.place_forget()
        restart.destroy()
        answer = 0
        answerbtn.configure(text = answer)
        question = random.randint(1, 50)
        questionbtn.configure(text = question)
        score = 0
        scorebtn.configure(text = "Score: "+str(score))
        scorebtn.place(x=10, y=540, height=50, width=100)
        x_object = random.randrange(150, 500, 50)
        y_object = 50
        object_value = random.randint(1, question)
        number.configure(text = object_value)
        number.place_forget()
        num_lives = 3
        lives.configure(text = "Lives left\n" + num_lives*"\U0001F49A")
        lives.place_forget()
        x_coord_of_basket = 6*50
        basket.place(x = x_coord_of_basket - 15, y = 20*25)
        basket_values_ls = []
        answer2btn.configure(text = ' '.join(map(str,basket_values_ls)))
        startingscreen.place(x = 20, y = 20)
        start2btn.place(x=10, y=340, height=50, width=100)
        start1btn.place(x=10, y=440, height=50, width=100)

    if running1 == True:
        running1 = False
        endscreen.configure(text = "Game Over!\n\nYour Score:\n" + str(score) + "\n")
        start1btn.place_forget()
        number.place_forget()
        questionbtn.place_forget()

    if running2 == True:
        running2 = False
        if len(final_ans_ls) == 0:
            endscreen.configure(text = "You Won!!\nCongrats!\n\n")
        else:
            endscreen.configure(text = "Game Over!\nYou coward >:(\n\n")
            
        start2btn.place_forget()
        object0.place_forget()
        object1.place_forget()
        object2.place_forget()
        question2btn.place_forget()
        answer2btn.place_forget()

    root.after(100, endscreen.place(x=10, y=10, height=600, width=600))
    restart = tk.Button(endscreen, text = "Restart", font = ('Courier, 20'), command = restart_game)
    root.after(150, restart.place(x=170, y=400, height=100, width=250))
    pass
    #end the game, reset all, display the score in the middle of the screen

# speed of falling object in mode 1:
speed = 100

#-------------------------------------------------------------------------------------------------------------------------
def minus_lives():
    global num_lives
    num_lives = num_lives - 1
    lives.configure(text = "Lives left\n" + num_lives*"\U0001F49A")

answer = 0
question = 10

score = 0
num_lives = 3

object_value = random.randint(1, 10)
x_object = 5*50
y_object = 50
number = tk.Label(root, text = object_value, font = ('Courier', 12))

def next_level():
    good_text = ["Your mother is proud of you!!", "Good Job!!!!!", "WOW!", "So smart", "slayyy", "+1 yay", "Amazing", "high score??", "full marks for you!", "yay! having fun?", "*slow clap*"]
    pick_good = random.randint(0, len(good_text)-1)
    message = tk.Label(root, text = good_text[pick_good], font = ('Comic Sans', 18), fg="#08E000")
    message.place(x = 130, y = 100)
    root.after(1000, message.place_forget)
    update()
    root.after(speed, falling1)

def failed_level():
    global num_lives
    bad_text = ["Greedy >:(", "Do better!", "Try again. but still -1 tho", "Again??", "Your mother is not proud", "Even my grandma can do better", "What was that??", "ouch..", "was that intentional?", "careful..", "going bootcamp?"]
    pick_bad = random.randint(0, len(bad_text)-1)
    message = tk.Label(root, text = bad_text[pick_bad], font = ('Comic Sans', 18), fg="#FF0000")
    if num_lives > 0:
        message.place(x = 130, y = 100)
    root.after(1000, message.place_forget)
    update()
    root.after(speed, falling1)

def update():
    def update_answer():
        #set answer back to 0
        global answer
        answer = 0
        answerbtn.configure(text = answer)

    def update_question():
        #creates a new question
        global question
        question = random.randint(1,50)
        questionbtn.configure(text = question)

    root.after(500, update_answer)
    root.after(500, update_question)

def falling1():
    global y_object, x_object, question, answer, number, object_value, score, lives
    #print(running1)
    if running1 == False or num_lives == 0:
        end_the_game()
        return
    elif running1 == True and num_lives != 0:
        y_object = y_object + 25
        number.place(x = x_object, y = y_object)
        questionbtn.place(x=200, y=10, height=30, width=200)
        lives.place(x = 10, y = 10)

        def make_new_object():
            global y_object, x_object, question, answer, number, object_value, score, num_lives
            y_object = 50
            x_object = random.randrange(150, 500, 50)
            if question - answer > 10:
                object_value = random.randint(10, (question-answer))
            elif question - answer < 4:
                object_value = random.randint(1, 4)
            else:
                object_value = random.randint(1, 10)
            number = tk.Label(root, text = object_value, font = ('Courier', 12))

        if y_object == 10*50 and x_object == x_coord_of_basket:
            answer = answer + object_value
            #make the number green
            collected_number = number
            collected_number = tk.Label(root, text = object_value, font = ('Courier', 12), fg="#08E000")
            collected_number.place(x = x_object, y = y_object)
            #destroys the number after it's been caught
            root.after(150, collected_number.place_forget)
            root.after(150, number.place_forget)
            #updates the display of the answer:
            answerbtn.configure(text = answer)

            if question == answer:
                next_level()
                green_answerbtn = tk.Button(root, text = answer, font=('Courier', 18), bg="#05FF00")
                green_answerbtn.place(x=250, y=540, height=50, width=100)
                green_questionbtn = tk.Button(root, text = question, font=('Courier', 18), bg="#05FF00")
                green_questionbtn.place(x=200, y=10, height=30, width=200)
                score = score + 1
                scorebtn.configure(text = "Score: "+str(score))
                root.after(250, green_answerbtn.destroy)
                root.after(250, green_questionbtn.destroy)

            elif question < answer:
                minus_lives()
                failed_level()
                red_answerbtn = tk.Button(root, text = answer, font=('Courier', 18), bg="#FF0000")
                red_answerbtn.place(x=250, y=540, height=50, width=100)
                red_questionbtn = tk.Button(root, text = question, font=('Courier', 18), bg="#FF0000")
                red_questionbtn.place(x=200, y=10, height=30, width=200)
#                 score = score - 1
#                 scorebtn.configure(text = "Score: "+str(score))
                root.after(250, red_answerbtn.destroy)
                root.after(250, red_questionbtn.destroy)

            else:
                make_new_object()
                root.after(speed, falling1)

        #looping falling function every speed ms
        elif question != answer and y_object <= 20*25:
            root.after(speed, falling1)

        #destroys the number if it is not caught after it fully falls down
        else:
            fallennumber = number
            fallennumber.place_forget()
            make_new_object()
            root.after(speed, falling1)

def running_switch_1():
    global running1
    startingscreen.place_forget()
    if running1 == True:
        end_the_game()
        return
    
    else:
        running1 = True
        answerbtn.place(x=250, y=540, height=50, width=100)
        start2btn.place_forget()
        root.after(250, falling1)

#----------------------------------------------------------------------------------------------------------------------

# Function that generates a list of random numbers for the game's final answer to achieve
def generate_final_ans():
    no_of_final_ans = 3 # To set the number of final answers
    max_final_ans = 99 # Set max integer for the final answer
    min_final_ans = 10 # Set min integer for the final answer
    final_answer_list = random.sample(range(min_final_ans,max_final_ans), no_of_final_ans) # To generate the final answers
    return final_answer_list

# Function to generate the numbers that drop
def generate_number_options(basket_values_ls):
    # Setting the number of options
    no_of_options = 3

    # Setting the range of options
    range_ls = [*range(1,10)]

    # Check if divide symbol is n basket_values_ls and return the appropriate values
    if chr(247) in basket_values_ls:

        # Reset range list
        range_ls = []

        # Add values to range_ls that can divide the first value
        for i in range(1,10):
            if basket_values_ls[0]%i == 0:
                range_ls.append(i)

        # If length of the range is less than the max option available, change the max option to the length of new range

        if len(range_ls) <= no_of_options-1:
            range_ls.append(0)

    options_list = random.sample(range_ls, no_of_options) # To generate the options

    #Shuffle the options so that they are not in ascending order
    random.shuffle(options_list)

    return options_list


def generate_operation(basket_values_ls):
    # List of math operations as the options to be dropped
    ls_operations = ['+', '-', 'x', chr(247)]

    count = 0

    for i in range(1,10):
        if basket_values_ls[0]%i == 0:
            count +=1

    if count < 3 or basket_values_ls[0] == 0:
        ls_operations = ['+', '-', 'x']

    # Shuffling the options so that they are not always in the same order
    random.shuffle(ls_operations)

    return ls_operations

# speed of falling object in mode 2:
speed2 = 150

# Create an empty list which would later be our values and operations in thr basket
basket_values_ls = []
#basket_values_ls[1] = operation_dict[basket_values_ls[1]]

# Create an empty list which would later be our final answers the players need to get
final_ans_ls = []

object_values_ls = [0, 0, 0]

x_coord_object_ls = [0, 0, 0]


def make_new_numop():
    global y_object, object0, object1, object2, object_values_ls, x_coord_object_ls, basket_values_ls
    # Dictionary for the operations to convert from string to proper python operators
    operation_dict = {'+':'+', '-':'-', 'x':'*', chr(247):'/'}
    y_object = 50
    # available x coordinates:
    choose_x_coord = [150, 200, 250, 300, 350, 400, 450]

    if len(basket_values_ls) == 0 or len(basket_values_ls) == 2:
        # runs the generate_number_options() function to produce the list of numbers to be dropped
        options = generate_number_options(basket_values_ls)
        object_values_ls = options.copy()
        #print(object_values_ls)

    # Drop the operations into the basket
    elif len(basket_values_ls) == 1:

        # runs generate_operation() function to produce the list of operations to be dropped
        operation = generate_operation(basket_values_ls)
        object_values_ls = operation.copy()

    object0 = tk.Label(root, text = object_values_ls[0], font = ('Courier', 12))
    object1 = tk.Label(root, text = object_values_ls[1], font = ('Courier', 12))
    object2 = tk.Label(root, text = object_values_ls[2], font = ('Courier', 12))

    x_coord_object_ls[0] = choose_x_coord.pop(random.randint(0, len(choose_x_coord)-1))
    x_coord_object_ls[1] = choose_x_coord.pop(random.randint(0, len(choose_x_coord)-1))
    x_coord_object_ls[2] = choose_x_coord.pop(random.randint(0, len(choose_x_coord)-1))

    object0.place(x = x_coord_object_ls[0], y = y_object)
    object1.place(x = x_coord_object_ls[1], y = y_object)
    object2.place(x = x_coord_object_ls[2], y = y_object)


def falling2():
    global running2, y_object, x_object, final_ans_ls, basket_values_ls, options
    global object0, object1, object2, object_values_ls, x_coord_object_ls
    if running2 == False or len(final_ans_ls) == 0:
        end_the_game()
        return
    elif running2 == True:
        # Continuously loop through this operation for the game to continue until the final answer has been achieved
        # Print the final answers that the players are trying to achieve
        question2btn.configure(text = str(final_ans_ls))
        question2btn.place(x=200, y=10, height=30, width=200)
        # Print the values and operations that is collected in the basket (already, update later)
        y_object = y_object + 25
        object0.place(x = x_coord_object_ls[0], y = y_object)
        object1.place(x = x_coord_object_ls[1], y = y_object)
        object2.place(x = x_coord_object_ls[2], y = y_object)

        # catching criteria
        # When object is caught into basket, append the values into the basket list
        if y_object == 20*25:
            if x_coord_of_basket == x_coord_object_ls[0]:
                basket_values_ls.append(object_values_ls[0])

                #make the object green
                collected_number = object0
                collected_number = tk.Label(root, text = object_values_ls[0], font = ('Courier', 12), fg="#08E000")
                collected_number.place(x = x_coord_object_ls[0], y = y_object)

                #destroys the number after it's been caught
                root.after(150, collected_number.place_forget)
                root.after(150, object0.place_forget)

            elif x_coord_of_basket == x_coord_object_ls[1]:
                basket_values_ls.append(object_values_ls[1])

                #make the object green
                collected_number = object1
                collected_number = tk.Label(root, text = object_values_ls[1], font = ('Courier', 12), fg="#08E000")
                collected_number.place(x = x_coord_object_ls[1], y = y_object)

                #destroys the number after it's been caught
                root.after(150, collected_number.place_forget)
                root.after(150, object0.place_forget)

            elif x_coord_of_basket == x_coord_object_ls[2]:
                basket_values_ls.append(object_values_ls[2])

                #make the object green
                collected_number = object1
                collected_number = tk.Label(root, text = object_values_ls[2], font = ('Courier', 12), fg="#08E000")
                collected_number.place(x = x_coord_object_ls[2], y = y_object)

                #destroys the number after it's been caught
                root.after(150, collected_number.place_forget)
                root.after(150, object0.place_forget)

            #updates the display of the answer:
            answer2btn.configure(text = ' '.join(map(str,basket_values_ls)))

            #destroys other numbers not caught after it fully falls down
            object0.place_forget()
            object1.place_forget()
            object2.place_forget()

            if len(basket_values_ls) == 3:
                operation_dict = {'+':'+', '-':'-', 'x':'*', chr(247):'/'}
                basket_values_ls[1] = operation_dict[basket_values_ls[1]]
                # Convert all in basket_values_ls to string, join them into one string and evaluate the string as a function
                basket_final_value = eval(''.join(map(str,basket_values_ls)))
                # change the number in basket values back into an int
                basket_values_ls = [int(basket_final_value)]
                # Update the basket with the new value
                answer2btn.configure(text = ' '.join(map(str,basket_values_ls)))

             # Check if the basket_final_value tallys with one of the list of final answers
                if basket_final_value in final_ans_ls:
                    # Remove the value from list if it tallys
                    final_ans_ls.remove(basket_final_value)
                    message = tk.Label(root, text = str(basket_final_value) + " is removed!", font = ('Comic Sans', 18), fg="#08E000")
                    message.place(x = 130, y = 100)
                    root.after(1000, message.place_forget)
                    question2btn.configure (text = str(final_ans_ls))
                    basket_values_ls = []
                    answer2btn.configure (text = ' '.join(map(str,basket_values_ls)))

            # If the list is empty, the player wins the game
            if len(final_ans_ls) == 0:
                #print("Congrats! You won the game!")
                end_the_game()
                return

            make_new_numop()
            root.after(speed2, falling2)

        #looping falling function every speed2 ms
        elif len(final_ans_ls) != 0 and y_object <= 20*25:
            root.after(speed2, falling2)

def running_switch_2():
    global running2, final_ans_ls
    startingscreen.place_forget()
    if running2 == True:
        end_the_game()
        return

    else:
        running2 = True
        start1btn.place_forget()
        scorebtn.place_forget()
        answer2btn.place(x=250, y=540, height=50, width=200)
        # Create a list of values which would be our final answers
        final_ans_ls = generate_final_ans()
        make_new_numop()

        root.after(250, falling2)

#------------------------------------------------------------------------------------------------------------------------

running1 = False
running2 = False

root.bind("<KeyPress>", move_basket) #binds a keypress to a shortcut/something happening inside the [root]

#------------------------------------------------------------------------------------------------------------------------

startingscreen = tk.Label(root, text="Hello! Press any of the Modes to start!\nPress Esc to end anytime!\n\nUse a / d or <- / ->\nto move the basket!\n\nAchieve the number(s) at the\ntop of the screen\n\nMode 1 = catch the numbers               \nMode 2 = catch both numbers and operators\n\n\n\n\n\n\n\n\n    basket:", font=('Courier', 16))
startingscreen.place(x = 20, y = 20)

start1btn = tk.Button(root, text = "Mode 1", font=('Courier', 18), command = running_switch_1)
start1btn.place(x=10, y=440, height=50, width=100)

questionbtn = tk.Button(root, text = question, font=('Courier', 18))

answerbtn = tk.Button(root, text = answer, font=('Courier', 18))

scorebtn = tk.Button(root, text = "Score: "+str(score), font=('Courier', 12))
scorebtn.place(x=10, y=540, height=50, width=100)

lives = tk.Label(root, text = "Lives left\n" + num_lives*"\U0001F49A", font = ('Courier', 18))

#----------------------------------------------------------------------------------------------------------------------

start2btn = tk.Button(root, text = "Mode 2", font=('Courier', 18), command = running_switch_2)
start2btn.place(x=10, y=340, height=50, width=100)

question2btn = tk.Button(root, text = str(final_ans_ls), font=('Courier', 18))

answer2btn = tk.Button(root, text = str(basket_values_ls),  font=('Courier', 14))

#----------------------------------------------------------------------------------------------------------------------

endscreen = tk.Label(root, text = "Game Over!\n\nYour Score:\n" + str(score) + "\n", font = ('Courier, 30'))

root.mainloop() #keeps the window open

```
